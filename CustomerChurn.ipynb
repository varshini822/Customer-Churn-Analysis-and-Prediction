```python
# ===============================
# Task 3: Customer Segmentation (Based on Your Dataset)
# ===============================

# --- Step 1: Account Length Segmentation ---
def account_group(length):
    if length < 100:
        return "Short (<100 days)"
    elif 100 <= length <= 200:
        return "Medium (100â€“200 days)"
    else:
        return "Long (>200 days)"

df["Account_Group"] = df["account_length"].apply(account_group)
```

```python
df["Usage_Group"] = pd.qcut(df["total_day_minutes"], q=3, labels=["Low", "Medium", "High"])
```

```python
def service_calls_group(calls):
    if calls <= 2:
        return "Low (0â€“2)"
    elif 3 <= calls <= 4:
        return "Medium (3â€“4)"
    else:
        return "High (5+)"

df["ServiceCalls_Group"] = df["customer_service_calls"].apply(service_calls_group)
```

```python
# --- Step 4: Churn Rate by Segments ---
seg_cols = ["Account_Group", "Usage_Group", "ServiceCalls_Group"]

for col in seg_cols:
    churn_rate = df.groupby(col)["churn"].mean() * 100
    print(f"\nChurn Rate by {col}:")
    print(churn_rate)

    plt.figure(figsize=(6,4))
    sns.barplot(x=churn_rate.index, y=churn_rate.values, palette="coolwarm")
    plt.title(f"Churn Rate by {col}")
    plt.ylabel("Churn Rate (%)")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# --- Step 5: Cross-Segment Analysis ---
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Account_Group", y="churn", hue="ServiceCalls_Group", palette="viridis")
plt.title("Churn by Account Length & Service Calls")
plt.ylabel("Churn Rate")
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Usage_Group", y="churn", hue="ServiceCalls_Group", palette="mako")
plt.title("Churn by Usage & Service Calls")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# ===============================
# Task 4: Predictive Modeling (Churn Prediction)
# ===============================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# --- Step 1: Drop non-numeric/segmentation columns ---
drop_cols = [
    "churn",
    "phone_number",
    "state",
    "tenure_group",
    "usage_category",
    "service_calls_category"
]

X = df.drop(columns=[col for col in drop_cols if col in df.columns])
y = df["churn"]
```

```python
# --- Step 2: Encode categorical variables ---
le = LabelEncoder()
if "international_plan" in X.columns:
    X["international_plan"] = le.fit_transform(X["international_plan"])
if "voice_mail_plan" in X.columns:
    X["voice_mail_plan"] = le.fit_transform(X["voice_mail_plan"])
```

```python
# --- Step 3: Ensure all features are numeric ---
X = X.apply(pd.to_numeric, errors="coerce")

# Drop any rows with NaN (if conversion failed)
X = X.fillna(0)
```

```python
# --- Step 4: Train-Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# --- Step 5: Feature Scaling ---
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# --- Step 6: Logistic Regression Model ---
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(X_train, y_train)

# --- Step 7: Predictions ---
y_pred = model.predict(X_test)
```

```python
# --- Step 8: Evaluation ---
acc = accuracy_score(y_test, y_pred)
print(f"\nâœ… Model Accuracy: {acc:.2%}")

print("\nðŸ“Š Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nðŸ“‘ Classification Report:")
print(classification_report(y_test, y_pred, digits=3))
```
